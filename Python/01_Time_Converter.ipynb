01_Time_Converter


```python
# Given number of minutes, convert it into human readable form."X hrs Y minutes" format
def convert_minutes(total_minutes):
    if total_minutes < 0:
        return "Invalid input"
    if total_minutes == 0:
        return "0 minutes"

    hours = total_minutes // 60
    mins = total_minutes % 60

    if hours == 0:
        return f"{mins} minutes"
    elif mins == 0:
        return f"{hours} hrs"
    else:
        return f"{hours} hrs {mins} minutes"

minutes=int(input("Enter minutes:"))
result =convert_minutes(minutes)
print(result)
```
